**Create Charts**

This notebook presents the code that is used to create the charts I demonstrate in my presentation.

The data used to create the charts is here: https://github.com/wel51x/wel51x.github.io/tree/master/Project1%20Data

```
# Imports and constants
import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

# hopefully optimize chart renderings
%config InlineBackend.figure_format='retina'

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.width', 500)

data_path = "https://raw.githubusercontent.com/wel51x/wel51x.github.io/master/Project1%20Data/"
```

```
# Print rental data info

apt_data = "CA%20Rent%202014-2018.csv"

df_apt = pd.read_csv(data_path + apt_data)

print("Rental Data Info")
print("shape:", df_apt.shape)
print(df_apt.describe(include='all'))

df_apt.set_index('county', inplace=True)
print("\nData:")
print(df_apt)
```

```
# Graph rentals

df_apt.plot(kind='bar', figsize=(20, 12),
            title="Monthly Rent for selected California counties 2014-2018")

plt.ylabel('Monthly Rent')
plt.show()
```

```
# Print house price data info

house_data = "CA%20Purchase%202014-2018.csv"

df_house = pd.read_csv(data_path + house_data)

print("House Price Data Info")
print("shape:", df_house.shape)

print(df_house.describe(include='all'))

df_house.set_index('county', inplace=True)
print("\nData:")
print(df_house)
```

```
# Graph house price data:

df_house.plot(kind='bar', figsize=(20, 12),
            title="House Prices for selected California counties 2014-2018")
#plt.hist(df_apt)
plt.ylabel('House Price')
plt.show()
```

```
# Print ratios of house price data to monthly rental costs info

ratio_data = "CA%20Purchase%20vs%20Rent%20Ratios%202014-2018.csv"

df_ratio = pd.read_csv(data_path + ratio_data)

print("Ratios of house price data to monthly rental costs Info")
print("shape:", df_ratio.shape)
print(df_ratio.describe(include='all'))

df_ratio.set_index('county', inplace=True)
print("\nData:")
print(df_ratio)
```

```
# Graph ratios

df_ratio.plot(kind='bar', figsize=(20, 12),
              title="Ratio of House Prices to Monthly Rent for selected California counties 2014-2018")

plt.show()
```

```
# Do same for 5 Bay Area counties

ratio_data = "CA%20Purchase%20vs%20Rent%20Ratios%202014-2018.csv"

df_ratio = pd.read_csv(data_path + ratio_data)

df5_counties = pd.DataFrame(df_ratio[(df_ratio['county'] == 'Alameda') |
                       (df_ratio['county'] == 'Marin') |
                       (df_ratio['county'] == 'San Francisco') |
                       (df_ratio['county'] == 'San Mateo') |
                       (df_ratio['county'] == 'Santa Clara') ],
                        columns = ["county", "2014", "2015", "2016", "2017", "2018"])

df5_counties.set_index('county', inplace=True)
print("\nData:")
print(df5_counties)
print("\n")

df5_counties.plot(kind='bar', figsize=(14, 8),ylim=(100, 500), rot = 0,
                  title="Ratio of House Prices to Monthly Rent for five Bay Area counties 2014-2018")
plt.show()
```

```
# Lastly do affordability
# First print info
price_to_agi_data = "CA%20House%20Price%20Ratio%20to%20AGI%20by%20County%202015.csv"

df_price_to_agi = pd.read_csv(data_path + price_to_agi_data)

print("Affordability Ratio data (house price to median income) Info")
print("shape:", df_price_to_agi.shape)
print(df_price_to_agi.describe(include='all'))

print("\nData:")
print(df_price_to_agi)
```

```
# Then graph

df_price_to_agi.plot('County', 'Ratio House Prices to AGI', kind='barh', legend='reverse',
                     figsize=(20, 12), xticks=df_price_to_agi.index, grid = True,
                     title="The Winston Ratio: House Prices vs Median Income")
plt.show()

```
